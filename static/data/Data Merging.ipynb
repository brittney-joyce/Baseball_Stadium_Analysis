```python
#import Dependacies 
import pandas as pd
import numpy as np
```

```python
#Storing file path locations
finance = "mlb_finances.csv"
wins = "Total_Wins.csv"
payroll = "Total_Payroll.csv"
attendance = "Total_Attendance.csv"
```

```python
#Creating Dataframes

finance_df = pd.read_csv(finance)
wins_df = pd.read_csv(wins)
payroll_df = pd.read_csv(payroll)
attendance_df = pd.read_csv(attendance)
```

```python
#test df
#finance_df
#wins_df
payroll_df
#attendance_df
```

```python
team_data1 = pd.merge(finance_df, wins_df, on="Team")
```

```python
team_data2 = pd.merge(team_data1, payroll_df, on="Team")
```

```python
Final_team_data = pd.merge(team_data2, attendance_df, on="Team")
```

```python
#Exporting DF to new CSV
Final_team_data.to_csv("Team_Data.csv", encoding = 'utf-8', index = False)
```
